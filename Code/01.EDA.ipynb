```python
#identify discrete and continuous features among numerical variables
#set threshold to define discrete features based on number of unique values
def identify_discrete_continuous(df, feas, threshold_unique=20):
    discrete_fea = []
    continuous_fea = []
    for fea in feas:
        temp = df[fea].nunique()
        #if the number of unique values is less than the threshold,consider it a discret feature
        if temp <= threshold_unique:
            discrete_fea.append(fea)
        else:
            continuous_fea.append(fea)
    return discrete_fea, continuous_fea

discrete_fea, continuous_fea = identify_discrete_continuous(data_train, numerical_fea)
```

```python
#define a function to sort the value in employmentLength column
def convert_employment_length(x):
    if pd.isnull(x):
        return 99
    if isinstance(x, str):
        x = x.strip().lower()
        if x == 'n/a' or x == 'nan':
            return 99
        elif '<' in x:
            return 0
        elif '10+' in x:
            return 10
        else:
            num = ''.join(c for c in x if c.isdigit())
            return int(num) if num else -1
    return -1
```

```python
#Visualizing the count of credit grade and employment length,
#distinguishing between defaulters and non-defaulters

data_train['employmentLength'] = data_train['employmentLength'].astype(str)

plt.figure(figsize=(10, 6))
sns.countplot(data=data_train, y='grade', hue='isDefault', palette={0: 'green', 1:'red'})
plt.title('Loan Default by Credit Grade')
plt.ylabel('Credit Grade')
plt.xlabel('Count')
plt.legend(title='Is Default')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
employment_order=sorted(data_train['employmentLength'].unique(), key=lambda x: convert_employment_length(x))
sns.countplot(data=data_train, y='employmentLength', hue='isDefault', order=employment_order, palette={0: 'green', 1:'red'})
plt.title('Loan Default by Employment Length')
plt.ylabel('Employment Length(Years)')
plt.xlabel('Count')
plt.legend(title='Is Default')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Total number of records(used to calculate percetage) & Total loan amount
total_users = len(data_train)
total_loan_amt = data_train['loanAmnt'].sum()

# Grouping and aggregating 'loanAmnt' feature
loan_amount_by_class = data_train.groupby('isDefault')['loanAmnt'].sum().reset_index()
loan_amount_by_class.columns = ['isDefault', 'totalLoanAmount']

# Create figure with 2 subplots
plt.figure(figsize=(14, 6))

# Subplot1: User Count Distribution(By Default Status)
plt.subplot(1, 2, 1)
count_plot = sns.countplot(x='isDefault', data=data_train, palette=['blue', 'red'])
count_plot.set_title("User Count by Default Status\n0 = Non-Default | 1 = Default", fontsize=14)
count_plot.set_xlabel("Default Status", fontsize=12)
count_plot.set_ylabel("User Count", fontsize=12)

# Add % labels on top
for p in count_plot.patches:
    height = p.get_height()
    percent = height / total_users * 100
    count_plot.text(p.get_x() + p.get_width()/2,
                    height + total_users * 0.01,
                    f"{percent:.2f}%",
                    ha='center',
                    fontsize=11)

# Subplot2: Loan Amount Distribution
plt.subplot(1, 2, 2)
amount_plot = sns.barplot(x='isDefault', y='totalLoanAmount', data=loan_amount_by_class, palette=['blue', 'red'])
amount_plot.set_title("Total Loan Amount by Default Status", fontsize=14)
amount_plot.set_xlabel("Default Status", fontsize=12)
amount_plot.set_ylabel("Total Loan Amount", fontsize=12)

# Add % label on top
for p in amount_plot.patches:
    height = p.get_height()
    percent = height / total_loan_amt * 100
    amount_plot.text(p.get_x() + p.get_width()/2,
                     height + total_loan_amt * 0.01,
                     f"{percent:.2f}%",
                     ha='center',
                     fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# Convert issueDate to datetime format
data_train['issueDate'] = pd.to_datetime(data_train['issueDate'], format='%Y-%m-%d')
data_testA['issueDate'] = pd.to_datetime(data_testA['issueDate'], format='%Y-%m-%d')

# Define a fixed baseline date
base_date = datetime.datetime.strptime('2007-06-01', '%Y-%m-%d')

# Create numerical date feature(days since base_date)
data_train['issueDateDT'] = (data_train['issueDate'] - base_date).dt.days
data_testA['issueDateDT'] = (data_testA['issueDate'] - base_date).dt.days

plt.figure(figsize=(10, 5))
plt.hist(data_train['issueDateDT'], bins=50, alpha=0.6, label='Train')
plt.hist(data_testA['issueDateDT'], bins=50, alpha=0.6, label='Test')
plt.xlabel('Days since 2007-06-01')
plt.ylabel('Frequence')
plt.title('Distribution of Issue Dates(Numerical)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Check for overlap in time ranges
train_min = data_train['issueDateDT'].min()
train_max = data_train['issueDateDT'].max()
test_min = data_testA['issueDateDT'].min()
test_max = data_testA['issueDateDT'].max()

print(f"Train issueDateDT range: {train_min} to {train_max}")
print(f"Test issueDateDT range: {test_min} to {test_max}")

overlap_start = max(train_min, test_min)
overlap_end = min(train_max, test_max)

if overlap_start <= overlap_end:
    print(f"Overlaping date range detected: {overlap_start} to {overlap_end}")
else:
    print("No overlap in issueDate between train and test sets")
```
